## **data visulaization using matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['SepalLengthCm'],df["SepalWidthCm"],label="length")
plt.xlabel("lenth")
plt.ylabel("flower")
plt.grid(True)
plt.legend()

```

```python
fig, axes = plt.subplots(2,2)
axes[0, 0].hist(df['SepalLengthCm'])
axes[0, 1].scatter(df['Id'], df['SepalLengthCm'])
axes[1, 0].plot(df['Id'], df['SepalWidthCm'])
axes[1, 1].plot(df['Id'], df['PetalLengthCm'])
```
